```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import accuracy_score
```

```python
x=covid.drop('COVID-19',axis=1)
y=covid['COVID-19']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3,random_state=180)
```

# Using KneighboursClassifier

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=20)
knn.fit(x_train, y_train)
y_pred = knn.predict(x_test)
acc_knn=(y_test==y_pred).sum()/len(y_test)
print("accuracy:",acc_knn)
```

# Estimation of severity using Knn

```python
c=pd.DataFrame({"predicted":y_pred,"actual:":y_test})
```

```python
# Now checking whether it is seviour or moderate or easy
print(c)
c['status']='ok'
```

```python
for i in c.index:
    so=0
    
    for j in covid.iloc[i, 0 : ]:
        
        if j==1:
            so+=1
  
   
    if covid['Breathing Problem'][i]==1 and (covid['Chronic Lung Disease'][i]==1 or covid['Fever'][i]==1):
        c['status'][i]='seviour'
        
    elif so>6 and covid['Fever'][i]==1:
        c['status'][i]='moderate'
    else:
        c['status'][i]='ok'
    
```
